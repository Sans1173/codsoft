```python
import pandas as pd
data = pd.read_csv("description.txt")
data.head(2)
```

```python
import pandas as pd
def load_data(file_path):

    with open(file_path, 'r', encoding='utf-8') as f:
        data1 = f.readlines()
    return[line.strip().split(':::') for line in data1]
```

```python
train_data = load_data('train_data.txt')
df_train = pd.DataFrame(train_data, columns = ['ID','Title', 'Genre', 'plot'])
test_data = load_data('test_data.txt')
df_test = pd.DataFrame(test_data, columns = ['ID','Title', 'plot'])
test_datasol = load_data('test_data_solution.txt')
df_test_sol = pd.DataFrame(test_datasol, columns = ['ID','Title', 'Genre', 'plot'])
print(train_data[0])
print(test_data[0])
print(test_datasol[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features = 10000, stop_words='english', ngram_range=(1,2))
x_train = vectorizer.fit_transform(df_train['plot'])
x_test = vectorizer.transform(df_test['plot'])
print(f"Training data:{x_train.shape}")
print(f"Testing data:{x_test.shape}")
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y_train = label.fit_transform(df_train['Genre'])
print(f"genre from training data:{label.classes_}")
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter= 300)
lr_model.fit(x_train, y_train)
y_pred = lr_model.predict(x_test)
predict_genre = label.inverse_transform(y_pred)
df_test['Predict_genre'] = predict_genre
```

```python
df_test[['Title', 'Predict_genre']]
```

```python
df_merged= pd.merge(df_test_sol[['ID', 'Genre']], df_test[['ID','Predict_genre']], on='ID')
df_merged
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(df_merged['Genre'], df_merged['Predict_genre'])
print(f"accuracy: {accuracy:.4f}")
```

```python
import warnings
warnings.filterwarnings('ignore')
print("classification_report:\n",classification_report(df_merged['Genre'], df_merged['Predict_genre']))
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_model = MultinomialNB()
NB_model.fit(x_train, y_train)
```

```python
y_pred_NB = NB_model.predict(x_test)
predict_genre_NB = label.inverse_transform(y_pred_NB)
df_test['Predict_genre_NB'] = predict_genre_NB
df_merged_NB= pd.merge(df_test_sol,df_test[['ID', 'Predict_genre_NB']], on='ID')
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_NB = accuracy_score(df_merged_NB['Genre'], df_merged_NB['Predict_genre_NB'])
print(f"accuracy: {accuracy_NB:.4f}")
print("classification_report:\n",classification_report(df_merged_NB['Genre'], df_merged_NB['Predict_genre_NB']))
```

```python
zoner_plot = [
    ' loves his life - his car, his apartment, his job','experiences as a professional wrestler',' tortured existence as one strange event'
]
test_data_tfidf =vectorizer.transform(zoner_plot)
y_pred = lr_model.predict(test_data_tfidf)
predict_genre= label.inverse_transform(y_pred)

y_pred_NB = NB_model.predict(test_data_tfidf)
predict_genre_NB = label.inverse_transform(y_pred_NB)

print("predict genre from plot")
print(f"logistic regression:{predict_genre}")
print(f"Naive Bayes:{predict_genre_NB}")
print()
for i , message in enumerate(zoner_plot):
    print(f"story : (message)")
    print(f"Status : \t Logistic Regression Prediction : {predict_genre[i]}")
    print(f"Status : \t Naive Bayes Prediction : {predict_genre_NB[i]}")
```
